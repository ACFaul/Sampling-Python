# Basic Box-Muller Transform

The basic Box-Muller transform generates random variables $x_1$ and $x_2$ drawn from the standard normal distribution by drawing two variables $y_1$ and $y_2$ from the uniform distribution on $(0,1)$ and letting
\begin{equation}
\begin{array}{rcl}
x_1 & = & \displaystyle \sqrt{-2 \log y_1} \cos (2 \pi y_2), \\[8pt]
x_2 & = & \displaystyle \sqrt{-2 \log y_1} \sin (2 \pi y_2).
\end{array}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def BoxMullerTrig(y1, y2):
  x1 = np.sqrt(-2*np.log(y1)) * np.cos (2* np.pi * y2)
  x2 = np.sqrt(-2*np.log(y1)) * np.sin (2* np.pi * y2)
  return (x1, x2)
```

```python
# Number of samples to be drawn.
N = 1000
# Initialising vector to hold samples
X = np.zeros((N,2))
# Generate samples.
for i in range(N):
  # Draw two samples from the uniform distribution.
  y1 = np.random.uniform()
  y2 = np.random.uniform()
  # Perform the basic Box-Muller transform.
  (X[i,0] ,X[i,1]) = BoxMullerTrig(y1, y2)
# Scatterplot.
plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
plt.show()
```
